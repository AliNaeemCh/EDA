# Zomato Dataset

### In EDA, we have the following objectives:

- Knowing about the dataset. Total columns, rows, datatypes etc.
- Handling missing values.
- Finding relatioships between features and their relevance to the target variable.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Increasing figure size
matplotlib.rcParams['figure.figsize'] = (12, 6)
```

```python
zomato = pd.read_csv('datasets/Zomato/zomato.csv', encoding='latin-1')
zomato.head()
```

```python
country_code = pd.read_excel('datasets/Zomato/Country-Code.xlsx')
country_code.head()
```

```python
df = pd.merge(zomato, country_code, on='Country Code', how='left')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

### Listing the missing values in all the columns

```python
df.isnull().sum()
```

### Listing the columns which have atleast 1 missing value

```python
[column for column in df.columns if df[column].isnull().sum()>0]
```

### Visualizing missing values

```python
sns.heatmap(df.isnull(), yticklabels=False, cmap='viridis')
```

### Listing number of records for each country

```python
country_vals = df['Country'].value_counts()
country_vals
```

### Creating a pie chart for top three countries

```python
country_names = country_vals.index
country_values = country_vals.values
plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
plt.show()
```

### Analyzing different rating counts

```python
ratings = df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings.head()
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
plt.show()
```

### Listing countries and their currencies

```python
df.groupby(['Country', 'Currency']).size().reset_index()
```

### Top 10 Cuisines

```python
cuisine_groups = df.groupby(['Cuisines'])['Aggregate rating'].agg(['mean']).rename(columns={'mean': 'Avg Rating'})
cuisine_groups.nlargest(10, 'Avg Rating')
```
